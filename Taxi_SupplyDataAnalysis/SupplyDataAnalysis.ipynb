# Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

# Reading and exploring data from CSV

```python
data_hour_act = pd.read_csv('Hourly_DriverActivity_1.csv - Hourly_DriverActivity_1.csv.csv', parse_dates=["Date"])
```

```python
data_search = pd.read_csv('Hourly_OverviewSearch_1.csv - Hourly_OverviewSearch_1.csv.csv', parse_dates=["Date"])
```

```python
data_search.head()
```

```python
data_search.describe()
```

# Merging two CSV files into one

```python
merged = data_hour_act.merge(data_search, on='Date')
```

```python
merged.head(24)
```

```python
merged.describe()
```

# Adding Time, Weekday and Week Number

```python
merged_hours = merged
```

```python
merged_hours['Date_only'] = [d.date() for d in merged_hours['Date']]
merged_hours['Time'] = [d.time() for d in merged_hours['Date']]
```

```python
merged_hours['Weekday'] = merged_hours['Date'].dt.weekday
merged_hours['Week_number'] = merged_hours['Date'].dt.week
```

```python
merged_hours.head(36)
```

```python
merged_hours.Time
```

# Grouping by Hour (Time)

```python
merged_hours_grouped = merged_hours.groupby('Time').mean()
```

```python
merged_hours_grouped.head()
```

```python
# total number of people who are looking for a car
merged_hours_grouped['People look for cars'] =\
    merged_hours_grouped['People saw 0 cars (unique)'] + merged_hours_grouped['People saw +1 cars (unique)']
```

```python
# over or undersupply (negative means that the number of drivers is less than the number of people)
merged_hours_grouped['People lack drivers'] =\
    merged_hours_grouped['Active drivers'] - merged_hours_grouped['People look for cars']
```

```python
merged_hours_grouped.describe()
```

```python
merged_hours_grouped.to_csv('24hours.csv')
```

# 24-hour curve of average supply and demand

```python
cols_plot = ['People look for cars', 'Active drivers']
merged_hours_grouped[cols_plot].plot(linewidth=1.5);
plt.suptitle('Average supply and demand comparison within 24-hours period', fontsize=16)
```

# Grouping by Weekday Number and Time

```python
merged_hours['People look for cars'] =\
    merged_hours['People saw 0 cars (unique)'] + merged_hours['People saw +1 cars (unique)']
```

```python
merged_hours['People lack drivers'] =\
    merged_hours['Active drivers'] - merged_hours['People look for cars']
```

```python
merged_grouped_week_time = merged_hours.groupby(['Weekday', 'Time']).mean()
```

```python
merged_grouped_week_time.head()
```

# 36 hours in a week, that are most undersupplied

Weekdays: 0 for Monday, 1 for Tuesday and so forth

```python
hours36 = merged_grouped_week_time.sort_values(by=['People lack drivers'], ascending=True).head(36)
hours36.to_csv('36hours.csv')
```

```python
# 8 EUR is drivers income from one ride
hours36['Avg driver inc per 1h rides'] =\
    hours36['Rides per online hour'] * 8
hours36
```

# Guaranteed hourly earnings

```python
hours36.describe()
# average income per 1h ride during these 36 hours is 6.99 EUR with 1.66 standard deviation
# which means we can safely guarantee 5.33 EUR (the difference between average and lower band of deviation)
# during these hours with condition that driver covers at least 12 hours (or 18)
# from these 36 hours (there are only 6 hours that have lower than 5.33 EUR earnings amount)
# this will improve our chances of being in plus

# in the end of the day it all boils down to balancing between our ability to increase supply and attract customers
```

# Guaranteed hourly earnings with increased supply

```python
hours36_extra_hours = hours36
hours36_extra_hours['Online (h)'] = hours36_extra_hours['Online (h)'] + 6
# we want to have 30+ hours in average, so by distributing these 6 hours over 36 hours will add 6 hours of all drivers in 1 hour
# currently average hours is 24.6
```

```python
# we may use this to recalculate our guaranteed income per hour in undersupplied 36 hours
hours36_extra_hours['Rides per online hour'] = hours36_extra_hours['Rides per online hour'] + 0.13
# 0.13 because this is the difference between ~ 30 online hours with 1.02 rides per hour (see 75%)
# and current average (24.6 and 0.87 respectively)
```

```python
# 8 EUR is drivers income from one ride
hours36_extra_hours['Avg driver inc per 1h rides'] =\
    hours36_extra_hours['Rides per online hour'] * 8
```

```python
hours36_extra_hours.describe()

# average income per 1h ride during these 36 hours assuming we increased supply is 8.02 EUR with 1.66 standard deviation
# which means we can safely guarantee 6.36 EUR (the difference between average and lower band of deviation)
```

# Coverage Ratio

```python
ratio = merged_grouped_week_time.sort_values(by=['Coverage Ratio (unique)'], ascending=True)
```

```python
ratio.corr()
```

```python
#ratio['Coverage Ratio (unique)'].corr(ratio['Online (h)'])
```

```python
ratio.head(10)
```

```python
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
ratio.plot(x='Online (h)', y='Coverage Ratio (unique)', style='o')  
plt.title('Title')  
plt.xlabel('Online (h)')  
plt.ylabel('Coverage Ratio (unique)')  
plt.show()

# the relationship between online hours and coverage ratio is not 100% linear,
# but still linear regression may give us a fairly good calculation
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(ratio['Online (h)'])

# there is a visable imbalance in distribution
# we might want to provide iniciatives for drivers with less hours
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(ratio['Coverage Ratio (unique)'])
```

```python
X = ratio['Online (h)'].values.reshape(-1,1)
y = ratio['Coverage Ratio (unique)'].values.reshape(-1,1)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)
```

```python
#This means that for every hour of change in online hours, the change in the Coverage Ratio is about 0.88%.
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
